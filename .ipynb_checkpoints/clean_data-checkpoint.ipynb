数据预处理

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.style.use('ggplot')
```

```python
df = pd.read_csv('air_data.csv')
df
```

```python
not_null_data = df.loc[df['SUM_YR_1'].notnull() & df['SUM_YR_2'].notnull(),:]
not_null_data
```

```python
index1 = not_null_data['SUM_YR_1'] != 0
index2 = not_null_data['SUM_YR_2'] != 0
index3 = (not_null_data['SEG_KM_SUM'] > 0) & (not_null_data['avg_discount'] != 0)
index4 = not_null_data['AGE'] > 100
clean_data = not_null_data[(index1 | index2) & index3 & ~index4]
clean_data
```

```python
clean_data.to_csv('clean_data.csv')
```
